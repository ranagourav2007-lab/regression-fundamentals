```python
import joblib
```

```python
LE=joblib.load("LablE.pkl")
LOGR=joblib.load("Logreg.pkl")
```

```python
['Gender', 'Study_Hours', 'Attendance', 'Previous_Score',
       'Final_Score' ]
```

```python
GENDER=input("write your gender -- Male,Female --")
Gender=LE.fit_transform([GENDER])[0]
print(Gender)

Study_Hours=int(input("write your Study_Hours"))
Attendance=int(input("write your Attendance"))
Previous_Score=int(input("write your Previous_Score"))
Final_Score=int(input("write your Final_Score"))
print(Gender,Study_Hours,Attendance,Previous_Score,Final_Score)
```

```python
prediction=LOGR.predict([[Gender,Study_Hours,Attendance,Previous_Score,Final_Score]])
if prediction == 1:
    print("pass")
else:
    print("fail")
```
